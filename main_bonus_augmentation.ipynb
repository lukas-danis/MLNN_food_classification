```python
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
from random import randint
```

```python
train_data_dir = 'data/train'
```

```python
datagen1 = ImageDataGenerator(
    rescale=1./255,
    brightness_range=[0.2,1]
)

it1 = datagen1.flow_from_directory(
    train_data_dir,
    color_mode="rgb",
    batch_size=1
)

datagen2 = ImageDataGenerator(
    rescale=1./255,
    width_shift_range=[-50,50]
)

it2 = datagen2.flow_from_directory(
    train_data_dir,
    color_mode="rgb",
    batch_size=1
)

datagen3 = ImageDataGenerator(
    rescale=1./255,
    vertical_flip=True
)

it3 = datagen3.flow_from_directory(
    train_data_dir,
    color_mode="rgb",
    batch_size=1
)
```

```python
my_dict_foods = it1.class_indices
rows = 1
columns = 4

#     brightness_range augmentation
fig = plt.figure(figsize=(20, 20))
for i in range(4):
    img, label = it1.next()
    lab=np.argmax(label)
    
    fig.add_subplot(rows, columns, i+1)
    plt.imshow(img[0])
    plt.axis('off')
    plt.title(list(my_dict_foods.keys())[list(my_dict_foods.values()).index(lab)])

fig.show()


#     width_shift_range augmentation
fig2 = plt.figure(figsize=(20, 20))

for i in range(4):
    img2, label2 = it2.next()
    lab2=np.argmax(label2)
    
    fig2.add_subplot(rows, columns, i+1)
    plt.imshow(img2[0])
    plt.axis('off')
    plt.title(list(my_dict_foods.keys())[list(my_dict_foods.values()).index(lab2)])
    
fig2.show()


#    shear_range augmentation 
fig3 = plt.figure(figsize=(20, 20))
for i in range(4):
    img3, label3 = it3.next()
    lab3=np.argmax(label3)
    
    fig3.add_subplot(rows, columns, i+1)
    plt.imshow(img3[0])
    plt.axis('off')
    plt.title(list(my_dict_foods.keys())[list(my_dict_foods.values()).index(lab3)])
    
fig3.show()
```
