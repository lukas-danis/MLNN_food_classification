```python
import hashlib
from matplotlib.pyplot import imread, imshow
import matplotlib.pyplot as plt
%matplotlib inline
import os
```

```python
os.chdir(r'C:\Users\danlu\Documents\suns_z3\data_duplicate\train')
file_list = next(os.walk('.'))[1]
```

```python
for dir_name in file_list:
    print('----Current directory----')
    print(dir_name)
    
    os.chdir(r'C:\Users\danlu\Documents\suns_z3\data_duplicate\train\\'+dir_name)
    
    print('\n----Number of files in directory----')
    files_list = os.listdir()
    print(len(files_list))
    
    duplicates = []
    hash_keys = dict()
    for index, filename in  enumerate(os.listdir('.')):
        if os.path.isfile(filename):
            with open(filename, 'rb') as f:
                filehash = hashlib.md5(f.read()).hexdigest()
            if filehash not in hash_keys: 
                hash_keys[filehash] = index
            else:
                duplicates.append((index,hash_keys[filehash]))

    print('\n----Found duplicates----')
    print(duplicates)
    
    for file_indexes in duplicates[:100]:
        try:
            plt.subplot(121),plt.imshow(imread(files_list[file_indexes[1]]))
            plt.title(file_indexes[1]), plt.xticks([]), plt.yticks([])

            plt.subplot(122),plt.imshow(imread(files_list[file_indexes[0]]))
            plt.title(str(file_indexes[0]) + ' duplicate'), plt.xticks([]), plt.yticks([])
            plt.show()

        except OSError as e:
            continue
    for index in duplicates:
        os.remove(files_list[index[0]])
    print('\n\n')
```
